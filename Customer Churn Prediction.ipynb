```python
import pandas as pd
data=pd.read_csv("C:/Users/User/OneDrive/Desktop/Mvie_dataset/archive (1)/churn.csv")
data.head()
data.info()
data.isnull().sum()
```

```python
#Feature Selection:
x=data.drop('Churn', axis=1)
y=data['Churn']
```

```python
#encoding:
encode=pd.get_dummies(x,drop_first=True)
```

```python
#Train test split:
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(encode,y,random_state=42,test_size=0.2)
```

```python
#feature scalling:
from sklearn.preprocessing import StandardScaler
ss= StandardScaler()
ss.fit_transform(x_train)
ss.transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
lr= LogisticRegression()
lr.fit(x_train,y_train)
rc=RandomForestClassifier()
rc.fit(x_train,y_train)
```

```python
y_pred1=lr.predict(x_test)
y_pred2=rc.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix, classification_report,accuracy_score
print("Classification Report(Logistic):",classification_report(y_test,y_pred1))
print("Classification Report(Random Forest):",classification_report(y_test,y_pred2))
```

```python
# Accuracy and comparisson:
from sklearn.metrics import accuracy_score

ac1 = accuracy_score(y_test, y_pred1) * 100
ac2 = accuracy_score(y_test, y_pred2) * 100

print('Accuracy Score (Logistic):', ac1, '%')
print('Accuracy Score (Random Forest):', ac2, '%')

if ac1 > ac2:
    print('Logistic gives more accuracy')
else:
    print('Random Forest gives more accuracy')
```
